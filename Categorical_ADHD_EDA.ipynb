```python
def get_category_column(data):
    '''
    Function to create a dataframe of category columns

    Inputs:
        data:
            DataFrame

    Returns:
        DataFrame of category columns
    '''

    # Select category columns
    category_columns = data.columns

    # Create a DataFrame to store information
    column_info = pd.DataFrame(
        columns=['Feature', 'Nunique', 'Percentage of missing values'])

    # Loop through each category column
    for column in category_columns:
        # Count unique values
        nunique = data[column].nunique()

        # Calculate percentage of NaN values
        percentage_nan = (data[column].isnull().sum() / len(data)) * 100

        # Append information to the DataFrame
        column_info = pd.concat([column_info, pd.DataFrame(
            {'Feature': column,
             'Nunique': nunique,
             'Percentage of missing values': percentage_nan},
            index=[0])],
            ignore_index=True)

    return column_info
```

```python
def plot_categorical_columns(df, column_name):
  """Plots a categorical column in a stacked bar chart relative to 'TARGET'.

  Args:
    df: The DataFrame containing the data.
    column_name: The name of the categorical column to plot.
  """

  cross_tab = pd.crosstab(df[column_name], df['ADHD_Outcome'])
  cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6))
  plt.title(f'Distribution of {column_name}')
  plt.xlabel(column_name)
  plt.ylabel('Count of values')
  plt.xticks(rotation=45)
  plt.legend(title='ADHD Outcome')
  plt.tight_layout()
  plt.show()


def calculate_adhd_proportion(df, column_name):
  """Calculates the proportion of target=1 (defaulters) for each value in a column.

  Args:
    df: The DataFrame containing the data.
    column_name: The name of the column to analyze.

  Returns:
      pandas.Series: A Series containing the proportion of defaulters for each value.
  """

  adhd_proportion = df.groupby(column_name)['ADHD_Outcome'].mean().rename('Percent %')
  return f'The ADHD proportion for each value in {adhd_proportion.round(4)*100}'
```

```python
occupation_mapping = {
    0: "Homemaker, stay at home parent",
    5: "Day laborer, janitor, house cleaner, farm worker, food counter sales, food preparation worker, busboy",
    10: "Garbage collector, short-order cook, cab driver, shoe sales, assembly line workers, masons, baggage porter",
    15: "Painter, skilled construction trade, sales clerk, truck driver, cook, sales counter or general office clerk",
    20: "Automobile mechanic, typist, locksmith, farmer, carpenter, receptionist, construction laborer, hairdresser",
    25: "Machinist, musician, bookkeeper, secretary, insurance sales, cabinet maker, personnel specialist, welder",
    30: "Supervisor, librarian, aircraft mechanic, artist and artisan, electrician, administrator, military enlisted personnel, buyer",
    35: "Nurse, skilled technician, medical technician, counselor, manager, police and fire personnel, financial manager, physical, occupational, speech therapist",
    40: "Mechanical, nuclear, and electrical engineer, educational administrator, veterinarian, military officer, elementary, high school and special education teacher",
    45: "Physician, attorney, professor, chemical and aerospace engineer, judge, CEO, senior manager, public official, psychologist, pharmacist, accountant"
}

# Grouping occupations by type
manual_labor = [0, 5, 10, 15, 20]
skilled_labor = [25, 30, 35]
intellectual_labor = [40, 45]
# Visualize the distribution of occupations by type
adhd_train['Barratt_Barratt_P1_Occ_Type'] = adhd_train['Barratt_Barratt_P1_Occ'].apply(
    lambda x: 'Manual Labor' if x in manual_labor else ('Skilled Labor' if x in skilled_labor else 'Intellectual Labor'))
print(calculate_adhd_proportion(adhd_train, 'Barratt_Barratt_P1_Occ_Type'))    
plot_categorical_columns(adhd_train, 'Barratt_Barratt_P1_Occ_Type')
```

```python
# Group families by parent's education and occupation type
family_groups = adhd_train.groupby([
    'Barratt_Barratt_P1_Edu', 
    'Barratt_Barratt_P2_Edu',
    'Barratt_Barratt_P1_Occ_Type', 
    'Barratt_Barratt_P2_Occ_Type'
])
family_counts = family_groups.size().reset_index(name='Family_Count')

# Display groups with the highest count first
print(family_counts.sort_values(by='Family_Count', ascending=False))
```

```python
# Tạo crosstab cho P1 với tỷ lệ ADHD
crosstab_p1 = pd.crosstab(
    [adhd_train['Barratt_Barratt_P1_Occ'], adhd_train['Barratt_Barratt_P1_Edu']],  
    adhd_train['ADHD_Outcome'],  
    normalize='index'  # Chuẩn hóa theo hàng (mỗi nhóm nghề nghiệp - giáo dục)
)

# Tạo crosstab cho P2 với tỷ lệ ADHD
crosstab_p2 = pd.crosstab(
    [adhd_train['Barratt_Barratt_P2_Occ'], adhd_train['Barratt_Barratt_P2_Edu']],  
    adhd_train['ADHD_Outcome'],  
    normalize='index'
)
# Lấy tỷ lệ bị ADHD (giả sử giá trị ADHD = 1 là bị mắc)
adhd_rate_p1 = crosstab_p1[1]  # Chọn cột ADHD = 1
adhd_rate_p2 = crosstab_p2[1]  # Chọn cột ADHD = 1

# Chuyển dữ liệu về dạng bảng phù hợp để vẽ heatmap
adhd_rate_p1 = adhd_rate_p1.unstack().fillna(0)  
adhd_rate_p2 = adhd_rate_p2.unstack().fillna(0)  

# Vẽ heatmap cho P1
plt.figure(figsize=(12, 8))
sns.heatmap(adhd_rate_p1, annot=True, cmap='YlOrRd', fmt='.2f')
plt.title('Tỷ lệ trẻ bị ADHD theo Nghề nghiệp và Giáo dục của P1')
plt.xlabel('P1 Education')
plt.ylabel('P1 Occupation')
plt.show()

# Vẽ heatmap cho P2
plt.figure(figsize=(12, 8))
sns.heatmap(adhd_rate_p2, annot=True, cmap='YlOrRd', fmt='.2f')
plt.title('Tỷ lệ trẻ bị ADHD theo Nghề nghiệp và Giáo dục của P2')
plt.xlabel('P2 Education')
plt.ylabel('P2 Occupation')
plt.show()
```

```python
# Tạo cột trung bình nghề nghiệp và học vấn của gia đình
adhd_train['Family_Occ'] = (adhd_train['Barratt_Barratt_P1_Occ'] + adhd_train['Barratt_Barratt_P2_Occ']) / 2
adhd_train['Family_Edu'] = (adhd_train['Barratt_Barratt_P1_Edu'] + adhd_train['Barratt_Barratt_P2_Edu']) / 2

# Nhóm nghề nghiệp thành bậc thấp, trung bình, cao
bins_occ = [0, 15, 30, 45]  # Nhóm: Thấp (0-15), Trung bình (20-30), Cao (35-45)
labels_occ = ['Low', 'Medium', 'High']
adhd_train['Occ_Group'] = pd.cut(adhd_train['Family_Occ'], bins=bins_occ, labels=labels_occ, include_lowest=True)

# Nhóm học vấn thành bậc thấp, trung bình, cao
bins_edu = [3, 12, 18, 21]  # Nhóm: Thấp (3-12), Trung bình (15-18), Cao (21)
labels_edu = ['Low', 'Medium', 'High']
adhd_train['Edu_Group'] = pd.cut(adhd_train['Family_Edu'], bins=bins_edu, labels=labels_edu, include_lowest=True)

# Tạo bảng crosstab với tỷ lệ ADHD theo nhóm nghề nghiệp và học vấn của gia đình
crosstab_family = pd.crosstab(
    [adhd_train['Occ_Group'], adhd_train['Edu_Group']],  # Nhóm gia đình
    adhd_train['ADHD_Outcome'],  
    normalize='index'  # Tính tỷ lệ theo hàng
)

# Chọn cột ADHD = 1 để lấy tỷ lệ mắc ADHD
adhd_rate_family = crosstab_family[1].unstack().fillna(0)  

plt.figure(figsize=(10, 6))
sns.heatmap(adhd_rate_family, annot=True, cmap='YlOrRd', fmt=".2f")
plt.title('Tỷ lệ trẻ mắc ADHD theo nhóm Nghề nghiệp và Học vấn của Gia đình')
plt.xlabel('Học vấn của Gia đình')
plt.ylabel('Nghề nghiệp của Gia đình')
plt.show()
```
